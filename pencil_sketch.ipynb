```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# read image using open cv

img=cv2.imread("mask.jpg")
```

```python
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis(False)
plt.show()
```

```python
#convert image to grey 

grey_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(grey_img)
plt.axis(False)
plt.show()
```

```python
#INVERT THE IMAGE

inv_img=255-grey_img
```

```python
# use gaussianblur to blur the image

blur_img=cv2.GaussianBlur(inv_img,(47,47),0)
plt.imshow(blur_img)
```

```python
#invert the blurred image

inv_blur_img=255-blur_img
```

```python
sketch=cv2.divide(grey_img,inv_blur_img,scale=256)
```

```python
cv2.imwrite('sketch.png', sketch)
```

```python
sketch_pic=cv2.cvtColor(sketch,cv2.COLOR_BGR2RGB)
plt.imshow(sketch_pic)
```
